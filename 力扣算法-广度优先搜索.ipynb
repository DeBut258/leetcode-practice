#### 广度优先搜索（Breadth-First Search, BFS）
**定义**：广度优先搜索是一种遍历或搜索树或图的方法，它从根节点开始，先访问当前节点的所有邻居节点，然后再依次访问这些邻居的邻居，逐层向外扩展，直到找到目标节点或遍历完整个图。

**通俗解释**：广度优先搜索就像波浪一样，从一个起点开始，向四周扩散，一层一层地向外延伸。

#### 例子
想象你在一个迷宫的入口，想找到出口。迷宫里每个房间都有若干门通向相邻的房间。你决定采用广度优先搜索的方法来找出口：

1. 你站在入口，首先查看从入口直接能到达的所有房间，并记录这些房间。
2. 然后你按顺序进入这些房间，每到一个新房间，查看从这个房间能到达的所有未访问的房间，再把这些房间记录下来。
3. 你不断重复这个过程，每次都优先处理当前层的所有房间，然后再进入下一层的房间。
4. 最终，你会找到出口，或者遍历完整个迷宫确认没有出口。

#### 广度优先搜索步骤
1. **起点**：从迷宫入口开始。
2. **第一层**：记录入口直接相邻的所有房间（这些房间就是第一层）。
3. **第二层**：依次进入第一层的每个房间，记录它们直接相邻的所有未访问房间（这些房间就是第二层）。
4. **重复**：继续这个过程，逐层向外扩展，直到找到出口或确认无路可走。

#### 总结
广度优先搜索就像在迷宫中一层一层地扩散，先查看离起点最近的所有房间，再查看更远的房间，确保不会遗漏任何可能的路径。这种方法保证了找到最短路径，因为它逐层扩展，不会跳过任何一个可能的路径。
